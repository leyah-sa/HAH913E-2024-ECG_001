### 1.2. Read the ecg file

```python
import numpy as np 

ecg = np.genfromtxt(fNameECG, delimiter=' ')
```

# 2. Plot the content of the `egc` variable 

```python
import matplotlib.pyplot as plt
import numpy as np

sampling_frequency = 1024  # Hz
gain = 0.000520  # V per unit

# Create a time vector based on the length of the ECG data and sampling frequency
time = np.arange(0, len(ecg)) / sampling_frequency  # Time in seconds

# Scale the ECG data by the gain
ecg_scaled = ecg * gain

plt.figure(figsize=(10, 6))
plt.plot(time, ecg_scaled[:, 0], label="Lead I")
plt.plot(time, ecg_scaled[:, 1], label="Lead II")
plt.plot(time, ecg_scaled[:, 2], label="Lead III")

plt.title("ECG Data for Leads I, II, III")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (mV)")
plt.legend()

plt.show()
```

## ECG Analysis

#### A normal ECG shows a regular pattern with distinct P waves, QRS complexes, and T waves.

#### There shouldn't be missing waves, irregular intervals. The consistency uniformity and regularity of the rythm should be checked also.


### Plot Lead I

```python
plt.figure(figsize=(10, 6))
plt.plot(time, ecg_scaled[:, 0], label="Lead I", color="blue")

plt.title("ECG Data for Lead I")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (mV)")
plt.legend()

plt.show()
```

### Plot Lead II

```python
plt.figure(figsize=(10, 6))
plt.plot(time, ecg_scaled[:, 1], label="Lead II", color="orange")

plt.title("ECG Data for Lead II")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (mV)")
plt.legend()

plt.show()
```

### Plot Lead III

```python
plt.figure(figsize=(10, 6))
plt.plot(time, ecg[:, 2], label="Lead III", color='green')

plt.title("ECG Data - Lead III")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (mV)")
plt.legend()

plt.show()
```
